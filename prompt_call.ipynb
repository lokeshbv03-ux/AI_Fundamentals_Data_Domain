```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
load_dotenv()

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
```

## Zero-Shot Prompting

```python
llm.invoke("Translate the following English sentence to French: 'How are you?'")
```

## One-Shot Prompting

```python
one_shot_prompt = (
    "Translate the following English sentence to French.\n"
    "Example: 'Good morning' -> 'Bonjour'\n"
    "Now, translate: 'How are you?'"
)
llm.invoke(one_shot_prompt)
```

## Few-Shot Prompting

```python
few_shot_prompt = (
    "Translate the following English sentences to French.\n"
    "Example 1: 'Good morning' -> 'Bonjour'\n"
    "Example 2: 'Thank you' -> 'Merci'\n"
    "Now, translate: 'How are you?'"
)
llm.invoke(few_shot_prompt)
```

## Chain-of-Thought Prompting

```python
cot_prompt = (
    "Solve the following math problem step by step.\n"
    "Question: If you have 10 apples and get 20 more, how many apples do you have?\n"
    "Example:\n"
    "Question: If you have 3 apples and get 2 more, how many apples do you have?\n"
    "Let's think step by step.\n"
    "Step 1: Start with 3 apples.\n"
    "Step 2: Get 2 more apples.\n"
    "Step 3: 3 + 2 = 5.\n"
    "Answer: 5 apples."
)
llm.invoke(cot_prompt)
```
